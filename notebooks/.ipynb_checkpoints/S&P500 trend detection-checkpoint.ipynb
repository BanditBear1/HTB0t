```python
import pandas as pd
import yfinance as yf

# Fetch historical data for S&P 500
ticker = "^GSPC"  # S&P 500 ticker symbol
data = yf.download(ticker, start="1980-01-01", end="2024-12-31")

# Calculate Exponential Moving Averages (EMAs)
data['EMA_5'] = data['Close'].ewm(span=5, adjust=False).mean()  # Fast EMA
data['EMA_20'] = data['Close'].ewm(span=10, adjust=False).mean()  # Slow EMA

# Calculate RSI (Relative Strength Index)
window_length = 14
delta = data['Close'].diff(1)
gain = (delta.where(delta > 0, 0)).rolling(window=window_length).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=window_length).mean()
rs = gain / loss
data['RSI'] = 100 - (100 / (1 + rs))

# Define conditions for RSI to make the trend stickier
overbought = 80
oversold = 20

# Identify the trend based on EMA crossovers and RSI confirmation
data['Trend'] = 'No Trend'
data.loc[(data['EMA_5'] > data['EMA_20']) & (data['RSI'] < overbought), 'Trend'] = 'Bullish'
data.loc[(data['EMA_5'] < data['EMA_20']) & (data['RSI'] > oversold), 'Trend'] = 'Bearish'

# Detect changes in trend
data['Trend_Change'] = data['Trend'].shift(1) != data['Trend']

# Filter for rows where the trend changes and add a column for the start date
trend_changes = data.loc[data['Trend_Change'] == True, ['Close', 'Trend']]
trend_changes['Start Date'] = trend_changes.index.strftime('%d %b %Y')

# Now we'll create a table that shows both the start and end date of each trend.
# Shift the start date to the next trend change to capture the end date of the current trend.
trend_changes['End Date'] = trend_changes['Start Date'].shift(-1)

# Remove the last row because it will not have a corresponding end date
trend_changes.dropna(inplace=True)

# Final table with the trend direction, start date, and end date
trend_table = trend_changes[['Trend', 'Start Date', 'End Date']].reset_index(drop=True)

# Rename the columns to match your desired output
trend_table.rename(columns={'Trend': 'Trend Direction'}, inplace=True)

# Display the dataframe in the console
print(trend_table)
```

```python
import matplotlib.pyplot as plt
# Plot S&P 500 and color code based on trend direction
plt.figure(figsize=(14, 7))

# Plot the closing price of the S&P 500
plt.plot(data.index, data['Close'], label='S&P 500 Close', color='black', linewidth=1.5)

# Highlight bullish periods
plt.fill_between(
    data.index,
    data['Close'],
    where=data['Trend'] == 'Bullish',
    color='green',
    alpha=0.3,
    label='Bullish'
)

# Highlight bearish periods
plt.fill_between(
    data.index,
    data['Close'],
    where=data['Trend'] == 'Bearish',
    color='red',
    alpha=0.3,
    label='Bearish'
)

# Add labels and title
plt.title('S&P 500 Price with Bullish and Bearish Trend Indicators', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.legend(loc='upper left')

# Display the plot
plt.grid(True)
plt.show()
```
